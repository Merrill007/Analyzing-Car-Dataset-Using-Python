## Data Cleaning

#### - Find all NULL values in the dataset. If there is any NULL value in any column, then fill it with the mean of that column.

```python
car.isnull()
```

```python
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

#### - Remove all the records where Weight is above 400

```python
car.head(2)
```

```python
car[~(car['Weight'] > 4000)]
```

#### - Applying function on a column - Increase all the values of the MPG_City column by 3

```python
car.head(2)
```

```python
car['MPG_City'] = car['MPG_City'].apply(lambda x:x+3)
```

```python
car
```

#### - What are the diferent types of make in our dataset. And what is the count of each Make in the data?

```python
car.head(2)
```

```python
# List of columns to fill missing values in with their respective means
columns_to_fill_with_mean = ['EngineSize', 'Horsepower', 'MPG_City', 'MPG_Highway', 'Weight', 'Wheelbase', 'Length']

# Filling in the missing values in the specified columns with their respective means
car[columns_to_fill_with_mean] = car[columns_to_fill_with_mean].fillna(car[columns_to_fill_with_mean].mean(numeric_only=True))
```

```python
car['Make'].value_counts()
```

```python
# i want to visualize this
# so i will import matplotlib

import matplotlib.pyplot as plt
```

```python
make_counts = car['Make'].value_counts()
```

```python
# Create a bar plot
plt.figure(figsize=(12, 6))  # Adjust the figure size as needed
bars = make_counts.plot(kind='bar', color='skyblue')
plt.title('Car Makes Distribution')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Adding data labels (count values) to the bars
for i, v in enumerate(make_counts):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

## Visuals

```python
import seaborn as sns

numerical_features = ['EngineSize', 'Cylinders', 'Horsepower', 'MPG_City', 'MPG_Highway', 'Weight', 'Wheelbase', 'Length']

# Calculate the correlation matrix
correlation_matrix = car[numerical_features].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))  # Adjust the figure size as needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

```python
# Correlation between EngineSize and Horsepower

x = car['EngineSize']
y = car['Horsepower']

# Create a scatter plot
plt.figure(figsize=(8, 6))  # Adjust the figure size as needed
plt.scatter(x, y, alpha=0.5, color='blue', edgecolors='black')
plt.title('Scatter Plot: Engine Size vs. Horsepower')
plt.xlabel('Engine Size')
plt.ylabel('Horsepower')
plt.grid(True)
plt.show()
```

```python
#distribution of 'car' with 'Make' and 'Type' columns

# Group the data by 'Make' and 'Type' and count the occurrences
grouped_data = car.groupby(['Make', 'Type']).size().unstack().fillna(0)

# Create a stacked bar chart
plt.figure(figsize=(14, 8))  # Adjust the figure size as needed
colors = plt.cm.Paired(range(len(grouped_data.index)))  # Color mapping for each 'Make'
grouped_data.plot(kind='bar', stacked=True, color=colors)
plt.title('Distribution of Car Types within Each Make')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.legend(title='Type', bbox_to_anchor=(1.05, 1), loc='upper left')  # Add legend

plt.tight_layout()
plt.show()
```

```python
# Top 10 cars with Highest Horsepower

# Sort the DF by 'Horsepower' in descending order to get the top N cars
N = 10  # Change N to the desired number of top cars
top_horsepower_cars = car[['Make', 'Horsepower']].sort_values(by='Horsepower', ascending=False).head(N)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
bars = plt.barh(top_horsepower_cars['Make'], top_horsepower_cars['Horsepower'], color='skyblue')
plt.title(f'Top {N} Cars with Highest Horsepower')
plt.xlabel('Horsepower')
plt.ylabel('Make')

# Add data labels (Horsepower values) to the bars
for bar, horsepower in zip(bars, top_horsepower_cars['Horsepower']):
    width = bar.get_width()
    plt.text(width + 10, bar.get_y() + bar.get_height() / 2, f'{horsepower:.1f}', va='center')

plt.tight_layout()
plt.gca().invert_yaxis()  # Invert the y-axis to display the highest horsepower at the top
plt.show()
```
